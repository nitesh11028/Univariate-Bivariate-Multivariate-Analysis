```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_sales= pd.read_csv('Sales_Data.csv')
df_sales.head(10)
```

```python
df1 = df_sales.Region.value_counts()
df1
```

```python
df_sales.Region.value_counts().plot.bar()
```

```python
sns.countplot(x ='Region',data=df_sales)
```

```python
plt.figure(figsize = (8,5))
sns.countplot(x ='Region',data=df_sales)
plt.title('Number of Count in Region')
plt.ylabel('Number of Orders')
plt.xticks(rotation=90)
plt.show()
```

```python
df_sales.Region.value_counts().plot.pie(autopct = '%.0f%%')
```

```python
a = pd.DataFrame(df_sales.Region.value_counts())
a
```

```python
plt.pie(a.Region,labels = a.index,autopct = '%.2f%%')
plt.show()
```

```python
colors = sns.color_palette('pastel')
plt.pie(a.Region,labels = a.index,autopct='%.0f%%',colors = colors)
plt.show()
```

```python
colors = sns.color_palette('muted')
plt.pie(a.Region,labels = a.index,autopct='%.3f%%',colors = colors)
plt.show()
```

```python
colors = sns.color_palette('deep')
plt.pie(a.Region,labels = a.index,autopct='%.4f%%',colors = colors)
plt.show()
```

```python
colors = sns.color_palette('dark')
plt.pie(a.Region,labels = a.index,autopct='%.6f%%',colors = colors)
plt.show()
```
